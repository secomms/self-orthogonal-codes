# Test showing that, using permutation matrix from LUP, we avoid unsolvable subsystems

Wait until a matrix with some singular first principal submatrix appears

```python
q = 11
k = 10

Fq = GF(q)

A = random_matrix(Fq, k, k)
while rank(A) < k:
    A = random_matrix(Fq, k, k)

ok = 1
for ell in range(k):
    sub_A = A[0:ell,0:ell]
    if rank(sub_A)<ell:
        ok = 0
        print("For ell = ",ell,"--> Rank Deficiency")
print("All subsystems ok?",ok==1)
    
```

We do LUP decomposition and then repeat the check on principal minors

```python
P, L, U = A.LU() #LUP decomposition
print("A == P*L*U?",A == P*L*U) #sanity check

new_A = P^-1*A #apply P

ok = 1
for ell in range(k):
    sub_A = new_A[0:ell,0:ell]
    if rank(sub_A)<ell:
        ok = 0
        print("For ell = ",ell,"--> Rank Deficiency")
print("All subsystems ok?",ok==1)
```

# Intensive tests

We now do several tests

```python
q = 2
k = 20
num_test = 10000


Fq = GF(q)
pr_fixed = 0
for i in range(1,num_test+1):
    
    ok = 1
    
    #generate matrix with some null first principal minor
    while ok:
        A = random_matrix(Fq, k, k)
        while rank(A) < k:
            A = random_matrix(Fq, k, k)

        for ell in range(k):
            sub_A = A[0:ell,0:ell]
            if rank(sub_A)<ell:
                ok = 0
                
    P, L, U = A.LU() #LUP decomposition

    new_A = P^-1*A #apply P

    ok = 1
    for ell in range(k):
        sub_A = new_A[0:ell,0:ell]
        if rank(sub_A)<ell:
            ok = 0
    
    pr_fixed += ok
    
    if (i%1000 == 0):
        print("Test #",i,", number of systems ok = ",N(pr_fixed/i*100),"%")
```
